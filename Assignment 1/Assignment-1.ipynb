```python
#import packages
import networkx as nx
import matplotlib.pyplot as plt
```

We need to establish Nodes and Edges in order to create the basic graph. Nodes are the poeple (Andre, Carol and etc...) in this case and Edges will be the connection between them. We are going to use "krackhardt_kite_graph" graph generator to create our Nodes and Edges.

```python
# Referencing https://networkx.github.io/documentation/networkx-1.10/reference/generated/networkx.generators.small.krackhardt_kite_graph.html
network=nx.krackhardt_kite_graph(create_using=None)
```

```python
#default network nodes
network.nodes()
```

```python
#default edges
network.edges()
```

```python
nx.draw(network, with_labels=1)
plt.show()
```

We created the network graph, now we will further customize the nodes and edges. 

```python
#customizing the labeling of the nodes (with people)
people = {0:'Andre', 1:"Beverly", 2:"Carol", 3:"Diane", 4:"Ed", 5:"Ike", 6:"Garth", 7:"Jane", 8:"Fernando", 9:"Heather"}
```

```python
# customizing the connection between the nodes
edges = [(0,1),(0,2),(0,3),(1,0),(1,3),(1,4),(2,0),(2,3),(2,8),(8,6),(8,3),(8,2),(6,4),(6,3),(6,8),(4,1),(4,3),(4,6),(9,8),
        (9,6),(9,5),(5,9),(5,7),(7,5)]
```

```python
#set up the graph
G = nx.Graph()
```

```python
#set up the edges
G.add_edges_from(edges)
```

```python
#label them with the nodes
G=nx.relabel_nodes(G,people)
```

```python
nx.draw(G, with_labels=1)
plt.show()
```

We can also create the graph with a seperate approach. The idea is the same, create nodes and edges.

```python
#create the nodes
nodes_2=['Andre','Beverly','Ed','Carol','Diane','Fernando','Ed','Garth','Heather','Ike','Jane']
#create the edges
edges_2=[('Andre','Beverly'),('Andre','Carol'),
        ('Andre','Diane'), ('Beverly','Andre'),
        ('Beverly','Diane'), ('Beverly','Ed'),
        ('Carol','Andre'), ('Carol','Diane'),
        ('Carol','Fernando'), ('Fernando','Garth'),
        ('Fernando','Diane'), ('Fernando','Carol'),
        ('Garth','Ed'), ('Garth','Diane'),
        ('Garth','fernando'), ('Garth','Heather'),
        ('Ed','Beverly'), ('Ed','Diane'),
        ('Ed','Garth'), ('Heather','Fernando'),
        ('Heather','Garth'), ('Heather','Ike'),
        ('Ike','Heather'), ('Ike','Jane'),
        ('Jane','Ike')]
```

```python
#set the graph
G_2=nx.Graph()
```

```python
#add the nodes
G_2.add_nodes_from(nodes_2)
```

```python
#add the edges
G_2.add_edges_from(edges_2)
```

```python
#draw and plot the graph
nx.draw(G_2, with_labels=True)
plt.show()
```
