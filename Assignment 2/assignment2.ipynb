## Data Collection

The Network Dataset we have selected is provided by Stanford's Large Network Dataset Collection. ( https://snap.stanford.edu/data/ ). We have selected Web-Google Network Data set where Nodes represents the web pages and directed edges represent the hyperlinks betweeen them. The dataset was part of the Google Programming Contest. The dataset can be found here: https://snap.stanford.edu/data/web-Google.html 

```python
# Load Packages
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
from pylab import rcParams
```

```python
df=pd.read_table('web-Google.txt', delim_whitespace=True, skiprows=4, names=['source', 'target'])
df.head()
```

```python
df.shape
```

The dataset has 5,105,039 edges. We will subset the dataset as suggested in order to work with just small subset of the nodes and edges.

```python
df=df[300:400]
df.shape
```

```python
df.head()
```

# Data Exploration

We have read the Network Dataset and we will further create our object and provide basic analysis. 

```python
# Creating the object
# reference - https://networkx.github.io/documentation/stable/reference/generated/networkx.convert_matrix.from_pandas_edgelist.html
G = nx.from_pandas_edgelist(df)
```

```python
# General information on the Graph
nx.info(G)
```

```python
# Number of Nodes
nx.number_of_nodes(G)
```

```python
nx.number_of_edges(G)
```

```python
# lets make it to start from node 30957 to node 565424
# Reference https://networkx.github.io/documentation/networkx-1.9/reference/generated/networkx.algorithms.shortest_paths.generic.shortest_path_length.html
nx.shortest_path_length(G, source=30957, target=565424, weight=None)
```

```python
# we can also not pick any source and target node
nx.shortest_path_length(G, source=None, target=None, weight=None)
```

```python
# we can look at the center of the graph
nx.center(G)
```

```python
# we can look at the graph diameter
nx.diameter(G)
```

```python
# calculate the shortest path (unweighted)
nx.all_pairs_shortest_path(G, cutoff=None)
```
